# 移动 `D:\下载` 下的全部 `.nii.gz` 文件到 `D:\FLARE22`

按顺序运行下面两个代码单元：

1. 扫描并查看 `D:\下载` 的全部下级目录，以及其中发现的 `.nii.gz` 文件。
2. 将发现的 `.nii.gz` 文件移动到 `D:\FLARE22`。移动后，文件的直接上级目录就是 `D:\FLARE22`。

如果 `D:\FLARE22` 中已经存在同名文件，移动代码会自动追加 `_1`、`_2` 等后缀，避免覆盖。

```python
from pathlib import Path

source_root = Path(r"D:\下载")
target_root = Path(r"D:\FLARE22")

if not source_root.exists():
    raise FileNotFoundError(f"源目录不存在: {source_root}")

all_dirs = sorted([p for p in source_root.rglob("*") if p.is_dir()], key=lambda p: str(p).lower())
nii_gz_files = sorted(source_root.rglob("*.nii.gz"), key=lambda p: str(p).lower())

print(f"源目录: {source_root}")
print(f"目标目录: {target_root}")
print(f"下级目录数量: {len(all_dirs)}")
print(f"发现 .nii.gz 文件数量: {len(nii_gz_files)}")

print("\n全部下级目录:")
for directory in all_dirs:
    print(directory)

print("\n发现的 .nii.gz 文件:")
for file_path in nii_gz_files:
    print(file_path)
```

```python
import shutil
from pathlib import Path

def unique_nii_gz_path(target_dir: Path, filename: str) -> Path:
    target_path = target_dir / filename
    if not target_path.exists():
        return target_path

    stem = filename[:-7] if filename.lower().endswith(".nii.gz") else Path(filename).stem
    suffix = ".nii.gz" if filename.lower().endswith(".nii.gz") else Path(filename).suffix

    counter = 1
    while True:
        candidate = target_dir / f"{stem}_{counter}{suffix}"
        if not candidate.exists():
            return candidate
        counter += 1

target_root.mkdir(parents=True, exist_ok=True)

moved = []
skipped = []

for source_path in nii_gz_files:
    if not source_path.exists():
        skipped.append((source_path, "源文件已不存在"))
        continue

    destination_path = unique_nii_gz_path(target_root, source_path.name)
    shutil.move(str(source_path), str(destination_path))
    moved.append((source_path, destination_path))

print(f"移动完成: {len(moved)} 个文件")
print(f"跳过: {len(skipped)} 个文件")

for source_path, destination_path in moved:
    print(f"MOVED: {source_path} -> {destination_path}")

for source_path, reason in skipped:
    print(f"SKIPPED: {source_path} ({reason})")

remaining = list(source_root.rglob("*.nii.gz"))
print(f"\n移动后 {source_root} 中剩余 .nii.gz 文件数量: {len(remaining)}")
print(f"移动后 {target_root} 中 .nii.gz 文件数量: {len(list(target_root.glob('*.nii.gz')))}")
```
